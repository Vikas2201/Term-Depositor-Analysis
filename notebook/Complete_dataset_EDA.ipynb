```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("without_imbalanced_dataset.csv")
data = data.drop(columns = ["Unnamed: 0"])
data
```

```python
data.shape
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
x_train = data.drop(columns = ["term_deposit_subscribed"])
y_train = data["term_deposit_subscribed"]
```

```python
rb = RobustScaler()
train_Scaled = rb.fit_transform(x_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand = RandomForestClassifier(random_state= 2)
```

```python
rand.fit(x_train,y_train)
```

```python
test_data = pd.read_csv("newtest.csv")
```

```python
test_data = test_data.drop(columns = ["Unnamed: 0"])
test_data.head()
```

```python
customer_id = test_data["customer_id"]
```

```python
test_data = test_data.drop(columns = ["customer_id"])
test_data
```

```python
test_data.shape
```

```python
Df = test_data.values
Df
```

```python
rand_lt = []
for i in range(len(Df)):
    scaled_data = rb.transform([Df[i]])
    prediction = rand.predict(scaled_data)
    rand_lt.append(prediction)
    print(prediction)
```

```python
pred_data = pd.DataFrame(rand_lt,columns = ["Predication"])
```

```python
pred_data.shape
```

```python
test_data["Predication"] = pred_data["Predication"]
test_data
```

```python
test_data=pd.concat([customer_id,test_data],axis = 1)
test_data.head()
```

```python
test_data = test_data[test_data["Predication"] == 1]
test_data
```

```python
test_data.shape
```

```python
df = test_data["customer_id"]
df
```
